```julia
using JchemoData, JLD2, CairoMakie
path_jdat = dirname(dirname(pathof(JchemoData)))
db = joinpath(path_jdat, "data/cassav.jld2")
@load db dat
pnames(dat)
```

```julia
X = dat.X
y = dat.Y.tbc
year = dat.Y.year
```

```julia
wl = names(X)
wl_num = parse.(Float64, wl)
```

```julia
plotsp(X, wl_num;
    xlabel ="Wavelength (nm)", ylabel = "Absorbance",
    title = "Cassava data").f
```
